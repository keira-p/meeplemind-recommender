### Early observations

**Data structure**

- Each set is connected via `BGGID`.
- `games`, `mechanics`, `themes`, `subcategories`, `publishers` and `rating_distribution` all have 21925 rows, ⭐️ one row per game.
- `user_ratings` is many rows per game.

**Game-level features**

- Games has 47 features describing the game, including:
  - ratings (`AvgRating`, `BayesAvgRating`)
  - difficulty level (`GameWeight`)
  - player counts and playtime, from makers and community
  - popularity metrics (e.g. `NumOwned`, `NumRatings`)
  - categorical flags to describe game type
  - ranking per game type
  - number of alternates, expansions and implementations
- This dataset includes several missing values which will need to be addressed.

**Categorical feature sets (multi-label)**

- `mechanics` (157), `themes` (217), `subcategories` (10) are represented as binary one-hot encoded features.
- 👉 A single game can have more then one mechanic, theme or subcategory. 
- Some games do not have these values recorded.

**Publishers**

- Publisher data is multi-label rather than single-label. 
- Most games have one publisher flagged, but many have multiple, likely reflecting regional editions, reprints, or multiple associated publishers.
- High dimensional (~1800 columns), less directly interpretable

**Ratings data**

- `ratings_distribution`: aggregated counts of ratings (0–10, 0.1 steps)
- `user_ratings`: individual user–game ratings
- Missing usernames present but not critical at this stage


## Decisions for next steps

- Use `BGGId` as the primary join key
- Treat `games` as the core table and base dataset
- Start with a game-level dataset for Project 1: *What makes a highly rated good game?*
  - Join in `mechanics`, `themes`, `subcategories` (multi-label binary feature sets)
  - Use `BayesAvgRating` as the main target variable (more robust that `AvgRating`)
- Keep initial scope focused:
  - Exclude `publishers` for now (high dimensional, less interpretable)
  - Exclude `user_ratings` (reserved for recommendation modelling)
  - Explore simple derived features e.g. number of mechanics, themes, subcategories per game
  - Revisit `user_ratings` in Project 2: *Game recommendation system*
